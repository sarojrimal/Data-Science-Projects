```python
job_ordinal = {"blue-collar":1, "management":2, "technician":3, "admin.":4, "services":5, "retired":6,
                "self-employed":7, "entrepreneur":8, "unemployed":9, "housemaid":10, "student":11, "unknown":12}
df["job_ordinal"] = df["job"].replace(job_ordinal)
df
```

# c) create a new column named age_category whose values should be based on the values in the age column, divide the values into 10ordinal numbers; 1 – 18 to 19, 2 – 20 to 25, 3 – 26 to 30, 4 – 30 to 35, 5 – 36 to 40, 6 – 41 to 49, 7 – 50 to 59, 8 – 60 to 69, 9 – 70 to 79, 10 – 80 and over

```python
df['age'].value_counts()
```

```python
def classes_def(x):
    if x == '18' or x=='19':
        return "1"
    elif x =='20' or x=='21' or x=='22' or x=='23' or x=='24' or x=='25':
        return "2"
df['age_category']=df['age'].apply(lambda x:classes_def(x))
target=df['age_category']

df['age_category'].value_counts()
```
